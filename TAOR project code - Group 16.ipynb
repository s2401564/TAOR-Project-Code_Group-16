# N = 6

```python
n = 5
# which means N = 6

with open('pcbouman-eur-TSP-D-Instances-8a6d795/singlecenter/singlecenter-11-n6.txt') as SC:
    n_lines = len(SC.readlines())
print(n_lines)

locations = []
with open('pcbouman-eur-TSP-D-Instances-8a6d795/singlecenter/singlecenter-11-n6.txt') as SC:
    n_lines = len(SC.readlines())
with open('pcbouman-eur-TSP-D-Instances-8a6d795/singlecenter/singlecenter-11-n6.txt') as SC:
    for i in range(1,n_lines+1):
        if i == 2:
            vehicle_speed = SC.readlines(i)[0]
            vehicle_speed = float(vehicle_speed)
        elif i == 4:
            drone_speed = SC.readlines(i)[0]
            drone_speed = float(drone_speed)
        elif i == 6:
            depot = SC.readlines(i)[0].split()
            locations.append(depot)
        elif i in range(8,(8+n)):
            locations.append(SC.readlines(i)[0].split())
        else:
            SC.readlines(i),i
            
# distance from i to j (Euclidean travel distances)
# initialise the distance matrix
distances = np.empty(shape=(n+1, n+1))
for i in range(n+1):
    for j in range(n+1):
        distances[i][j] = math.sqrt((float(locations[j][1]) - float(locations[i][1]))**2 + 
                        (float(locations[j][0]) - float(locations[i][0]))**2)
```

# Input parameters

```python
# The total number of nodes in the network
N = n+1

# Set of all nodes in the network
number_of_nodes = N
V = range(number_of_nodes)

# Set of customer nodes, which excludes the central depot
number_of_customer_nodes = N-1
V_star = range(1,N)

# Vehicle travel time from i to j
#t = [(i,j) for i in V for j in V if i !=j]
t =  np.empty(shape=(N, N))
for i in range(N):
    for j in range(N):
        t[i][j] = distances[i][j]/vehicle_speed
        
# Drone travel time from i to j
f =  np.empty(shape=(N, N))
for i in range(N):
    for j in range(N):
        f[i][j] = distances[i][j]/drone_speed
        
# D_max is longer than the duration of a feasible TSP-D tour
D_max = 0
for i in V:
    for j in V:
        D_max += distances[i][j]/1

# Maximum flight duration for the drone to make a single delivery then return to the vehicle/30 minutes
F = 30

# A “large enough” scalar
A = [0]
for i in V_star:
    for j in V_star:
        if (f[i][j] + f[j][0] <= F):
            A.append(min(f[j][0],t[j][0]))
A = min(A)
M = D_max - A
```

```python
V  = [i for i in V]


def sub_lists (l):
    lists = []
    for i in range(len(l) + 1):
        for j in range(i):
            lists.append(l[j: i])
    return lists

S = sub_lists(V)
```

# Decision variables

```python
# the vehicle travels from i to j
#x = np.array([xp.var(vartype = xp.binary, name='x_{0}_{1}'.format(i+1,j+1))
#                    for i in V for j in V if i != j]).reshape(number_of_nodes,number_of_nodes)
x = np.array([xp.var(vartype = xp.binary, name='x_{0}_{1}'.format(i+1,j+1))
                    for i in V for j in V]).reshape(number_of_nodes,number_of_nodes)

# the drone travels from i to j
y = np.array([xp.var(vartype = xp.binary, name='y_{0}_{1}'.format(i+1,j+1))
                    for i in V for j in V]).reshape(number_of_nodes,number_of_nodes)

# the drone flies from i to j
z = np.array([xp.var(name='z_{0}_{1}'.format(i+1,j+1))
                    for i in V for j in V], dtype=xp.npvar).reshape(number_of_nodes,number_of_nodes)

# the vehicle visits node j
w = np.array([xp.var(name='w_{0}'.format(j+1))
                    for j in V], dtype=xp.npvar)

# the arrival time of the delivery carrier at node j
a = np.array([xp.var(name='a_{0}'.format(j+1))
                    for j in V], dtype=xp.npvar)

# the departure time of the delivery carrier at node j
d = np.array([xp.var(name='d_{0}'.format(j+1))
                    for j in V], dtype=xp.npvar)

# the time spent by the vehicle waiting for the drone at node j
e = np.array([xp.var(name='e_{0}'.format(j+1))
                    for j in V], dtype=xp.npvar)
```

# Constraints

```python
#(1b) vehicle departs the depot for exactly once
prob.addConstraint(xp.Sum(x[0][j] for j in V_star) == 1)
```

```python
#(1c) node i visited by vehicle
prob.addConstraint(xp.Sum(x[i][j] for j in V) == w[i] for i in V)
```

```python
#(1d) vehicle departs the depot for exactly once
prob.addConstraint(xp.Sum(x[i][j] for j in V) - xp.Sum(x[j][i] for j in V) == 0 for i in V)
```

```python
#(1e) drone departs the depot for exactly once
prob.addConstraint(xp.Sum(y[0][j] for j in V_star) == 1)
```

```python
#(1f) drone departs the depot for exactly once
prob.addConstraint(xp.Sum(y[i][j] for j in V) - xp.Sum(y[j][i] for j in V) == 0 for i in V)
```

```python
#(1g) enforce visited by drone if absence vehicle
prob.addConstraint(xp.Sum(y[i][j] for j in V) >= 1 - w[i] for i in V)
prob.addConstraint(xp.Sum(y[i][j] for j in V) <= 1 for i in V)
```

```python
#(1h) at least one of i or j by vehicle
prob.addConstraint(w[i] + w[j] >= y[i][j] for i in V for j in V if i!= j)
```

```python
#(1i) drone from i to j and served by vehicle, vehicle must travel
prob.addConstraint(y[i][j] + w[i] + w[j] <= x[i][j] + 2 for i in V for j in V if i!= j)
```

```python
#(1j) drone fly from i to j if not transported aboard the vehicle
prob.addConstraint(y[i][j] - x[i][j] <= z[i][j] for i in V for j in V if i!= j)
prob.addConstraint(z[i][j] <= 1 - x[i][j] for i in V for j in V if i!= j)
```

```python
#(1k) if drone fly from i to j, must travel
prob.addConstraint(z[i][j] <= y[i][j] for i in V for j in V if i!= j)
```

```python
#(1l) given flight, drone travel at most a duration F before meeting vehicle
prob.addConstraint(xp.Sum(f[i][j]*z[i][j] for i in V) + xp.Sum(f[j][k]*z[j][k] for k in V) <= 
                   F + M*w[j] for j in V_star)
```

```python
#(1m)(1n)(1o)(1p) arrival and departure times from (MTZ)
prob.addConstraint(a[j] >= d[i] + t[i][j] - M*(1 - x[i][j]) for i in V for j in V if i!= j)
prob.addConstraint(d[j] >= a[j] for j in V_star)
prob.addConstraint(a[j] >= d[i] + f[i][j] - M*(1 - z[i][j]) - M*w[j] for i in V for j in V if i!= j)
prob.addConstraint(d[j] >= d[i] + f[i][j] - M*(1 - z[i][j]) for i in V for j in V_star)
```

```python
#(1q) vehicle waiting = (actual - expected)departure time
prob.addConstraint(e[j] >= d[j] - (d[i] + t[i][j]) - M*(1 - x[i][j]) for i in V for j in V_star if i!= j)
```

```python
#(1r) drone later than vehicle, waiting equal to difference
prob.addConstraint(e[0] >= (d[j] + f[j][0]) - (d[i] + t[i][0]) - M*(1 - x[i][0]) 
                   for i in V_star for j in V_star if i!= j)
```

```python
#(1s) latest arrival time of either carrier
prob.addConstraint(a[0] <= xp.Sum(xp.Sum(t[i][j]*x[i][j] for j in V) for i in V) + 
                   xp.Sum(e[k] for k in V))
```

```python
#(1t)
prob.addConstraint(a[j] >= 0 for j in V)
prob.addConstraint(d[j] >= 0 for j in V)
prob.addConstraint(w[j] >= 0 for j in V)
prob.addConstraint(z[i][j] >= 0 for i in V for j in V)
prob.addConstraint(e[j] >= 0 for j in V)
prob.addConstraint(w[j] <= 1 for j in V)
```

```python
#(1u)
prob.addConstraint(a[j] <= M for j in V)
prob.addConstraint(d[j] <= M for j in V)
prob.addConstraint(e[j] <= 
                   max(max((f[i][k] + f[k][j] - t[i][j]) 
                           for i in V for k in V if (f[i][k] + f[k][j] <= F)),0) 
                   for j in V)
```

```python
prob.addConstraint(x[i][j] == 0 for i in V for j in V if i==j)
prob.addConstraint(y[i][j] == 0 for i in V for j in V if i==j)
```

```python
# Step 1: Solve an MIP relaxation of Model TSPD2 -----(TSPD2_1), 
# in which x-variables are binary, while the y-variables are continuous; 
# let the resulting solution be (X2_1,W2_1):
```

```python
def step1():
    prob2_1.setControl('maxtime',30)
    prob2_1.solve()
    X2_1 = np.empty([N,N])
    for i in V:
        for j in V:
            X2_1[i][j] = prob2_1.getSolution(x2_1[i][j])
    W2_1 = np.empty(N)
    for i in V:
        W2_1[i] = prob2_1.getSolution(w2_1[i])
        
    return(W2_1)
```

```python
# Step 2: For each W2_1 = 1, set w = 1. 
# Using the DoubleGreedy procedure, generate an initial feasible solution using the fixed w-variables. 
# Let Zwarm be the solution generated with the fixed w-values, and let Dmax = Zwarm
```

```python
def step2(W2_1):
    # W for doublegreedy
    W_DG = [0]*N
    W_DG[0] = 1
    for i in V_star:
        if W2_1[i] > 0.99:
            W_DG[i] = 1
    # the nodes visited by the vehicle
    vehicle = []
    route = [0]
    Z_warm = 0
    route_time = []
    # the nodes must be visited by the drone
    drone = []
    for i in V_star:
        if W2_1[i] > 0.99:
            vehicle.append(i)
        else:
            drone.append(i)
    vehicle1 = vehicle.copy()

    # find the first cumstor node
    t_arc = t[route[-1]][vehicle[0]]
    last_node = vehicle[0]
    for j in vehicle:
        if t[0][j] < t_arc:
            last_node = j
            t_arc = t[0][j]
    route.append(last_node)
    vehicle.remove(last_node)
    Z_warm += t_arc
    route_time.append(t_arc)

    for i in range(len(vehicle1)-1):
        t_arc = t[route[-1]][vehicle[0]]
        last_node = vehicle[0]
        for j in vehicle:
            if t[route[-1]][j] < t_arc:
                last_node = j
                t_arc = t[route[-1]][j]
        route.append(last_node)
        vehicle.remove(last_node)
        Z_warm += t_arc
        route_time.append(t_arc)

    route.append(0)
    Z_warm += t[route[-2]][0]
    route_time.append(t[route[-2]][0])
    
    drone1 = drone.copy()
    drone_complete = 0
    for i in route:
        if drone == []:
            break
        f_arc = f[i][drone[0]]
        drone_node = drone[0]
        f_arc1 = Z_warm
        meet_node = 0
        for j in drone:
            if f[i][j] < f_arc:
                f_arc = f[i][j]
                drone_node = j
        for k in route[i:len(route)-3*(len(drone)-1)]:
            if f[drone_node][k] < f_arc1:
                f_arc1 = f[drone_node][k]
                meet_node = k
        drone_time = (f_arc + f_arc1)/2
        vehicle_time = sum(route_time[route.index(i):(route.index(meet_node)+1)])
        if drone_time > vehicle_time:
            Z_warm += (drone_time - vehicle_time)
        i = meet_node
        drone.remove(drone_node)
        drone_complete += 1
        if drone_complete >= len(drone1):
            break
            
    D_max = Z_warm
    return(D_max)
```
